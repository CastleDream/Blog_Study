## 实验1

```python
def relu_op1(x):
    y1 = max(0,x)
    y2 = 0.5*y1
    y3 = max(0,y2)
    y4 = -2*y3
    y5 = y4+1
    y6 = max(0,y5)
    y7 = 2*(y6+2)
    y8 = max(0,y7)
    return y8
```

```python
import numpy as np
x_list = np.arange(-5,10).tolist()
print(x_list[:4])
y_list = [relu_op1(x) for x in x_list]
print(y_list[:4])

import matplotlib.pyplot as plt
plt.plot(x_list,y_list)
```

## 实验2

```python
def relu_op2(x):
    y0 = 3*x+0.01
    relu1 = max(0,y0)
    y1 = 0.5*relu1+1.4
    relu2 = max(0,y1)
    y2 = -2*relu2+3
    relu3 = max(0,y2)
    y3 = 2*relu3-1
    return y3
```

```python
import numpy as np
x_list = np.arange(-5,10).tolist()
print(x_list[:4])
y_list = [relu_op2(x) for x in x_list]
print(y_list[:4])

import matplotlib.pyplot as plt
plt.plot(x_list,y_list)
```

## 实验3

考虑一个小范围的，x$\in$ [-6,10]

```python
def relu_op2(x):
    """
    x∈[-6,10]
    """
    y0 = 0.5*x+2  # x∈[-6,10]，y0∈[-1,7]
    relu1 = max(0,y0) # relu1∈[0,7]
    y1 = relu1-4 # y1∈[-4,3]
    relu2 = max(0,y1) # relu1∈[0,3]
    y2 = -2*relu2+3 #y2∈[-3,3]
    relu3 = max(0,y2) # relu3∈[0,3]
    y3 = 2*relu3-7
    return y3
import numpy as np
x_list = np.arange(-5,10).tolist()
print(x_list[:4])
y_list = [relu_op2(x) for x in x_list]
print(y_list[:4])

import matplotlib.pyplot as plt
plt.plot(x_list,y_list)
```

```python
def relu_op2(x):
    """
    x∈[-6,10]
    """
    y0 = 0.5*x+2  # x∈[-6,10]，y0∈[-1,7]
    relu1 = max(0,y0) # relu1∈[0,7]
    y1 = 0.5*relu1+1.4
    relu2 = max(0,y1)
    y2 = -2*relu2+3
    relu3 = max(0,y2)
    y3 = 2*relu3-1
    return [y0,relu1,y1,relu2,y2,relu3,y3]

import numpy as np
x_list = np.arange(-6,10).tolist()
print(x_list[:4])

for x in x_list:
    y0,relu1,y1,relu2,y2,relu3,y3 = relu_op2(x)
    print(y0,relu1,y1,relu2,y2,relu3,y3)
    break

# 这样得到的是每个x对应的每组值，我需要的是y0一类（一个numpy），relu1一类（一个numpy）
y_list = [relu_op2(x) for x in x_list]
print(y_list[:4])

y_np = np.array(y_list)
y_transpose = np.transpose(y_np)
y_list = y_transpose.tolist()
print(f"transpose y_list: {y_list[:4]}")

import matplotlib.pyplot as plt
for y_single in y_list:
    plt.plot(x_list,y_single)
```
