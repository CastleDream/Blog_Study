### 异常处理

#### try-except

try-except可以对异常进行捕捉处理，例如

```python
try:
    num1=int(input('Enter the first number:'))
    num2=int(input('Enter the second number:'))
    print(num1/num2)
except ValueError:
    print("Please input a digit")
```

+ 上面这个例子是 如果输入一个非数值的数值，则异常处理方式是提示 输入一个数字，但是因为没有使用循环，所以程序直接在这里停住了
+ 另外，这样写一个程序只能捕捉一种异常，还可以捕捉多种异常

```python
try:
    num1=int(input('Enter the first number:'))
    num2=int(input('Enter the second number:'))
    print(num1/num2)
except ValueError as err:
    print("Please input a digit")
    print(err)
```

* 这样不仅打印出自己自定义的异常处理方式，也打印出默认的异常类给出的提示信息
* except exception as e 是很常用的一种方式

```python
try:
    num1=int(input('Enter the first number:'))
    num2=int(input('Enter the second number:'))
    print(num1/num2)
except ValueError as err:
    print("Please input a digit")
    print(err)
except ZeroDivisionError:
    print('The second number can\'t be zero!')
```

+ 使用多个except语句来捕捉多个异常
+ 或者使用except块来捕捉多个异常，如下例：

```python
try:
    num1=int(input('Enter the first number:'))
    num2=int(input('Enter the second number:'))
    print(num1/num2)
except (ValueError,ZeroDivisionError) as err:
    print("Invalid input")
    print(err)
```

所以也可以这样写， as err此时就成了一个list形式的错误，汇总了之前所有的，写程序的人真厉害，考虑到了扩展，666

```python
try:
    num1=int(input('Enter the first number:'))
    num2=int(input('Enter the second number:'))
    print(num1/num2)
except:
    print("Something went wrong")
```

使用空的except子句就可以捕捉所有的异常，但是这样的提示似乎不太好，可以使用下面这样的方式:
* except Exception as e:
+ 这样更明确，友好 

```python
try:
    num1=int(input('Enter the first number:'))
    num2=int(input('Enter the second number:'))
    print(num1/num2)
except Exception as e:
    print(e)
```

#### else子句

```python
try:
    num1=int(input('Enter the first number:'))
    num2=int(input('Enter the second number:'))
    print(num1/num2)
except:
    print("Something went wrong")
else:
    print("Everything is Ok")
```

还可以使用else语句，表明没有异常发生，也就是try-except-else 有异常发生，则执行except子句 没有异常发生，就执行else

但是上述例子，都是异常发生后给出提示信息，但是程序也因此终止，想要在发生异常后还可以继续输入，直到输入正确为止，可以使用循环

```python
while True:
    try:
        num1=int(input('Enter the first number:'))
        num2=int(input('Enter the second number:'))
        print(num1/num2)
        break
    except:
        print("Something went wrong")
    else:
        print("Everything is Ok")
```

+ 使用循环，直到try执行正确，才能执行break，循环才会终止
+ break的位置不一定在try语句块里，也可以在else或者except里，视情况而定

#### finally子句

不管是否发生异常，都会执行finally，不同于except和else是或者的关系，这个是不管如何都会执行的

#### 上下文管理器（Context Manager）和with语句

```python
try:
    f=open('test_tweets.csv',encoding='utf8')
    lista=[]
    for line in f:
        lista.append(line)
    print(lista[1:4])
except:
    print('Cannot open the file!')
finally:
    f.close()
```

+ 但是，假设文件已经不能正常打开了，那么上面的f其实没有获取到值，所以f.close（）仍然会发生异常
+ 所以，python中提供了上下文管理器，通过上下文管理器，可以定义和控制代码执行前的准备动作以及执行后的收尾动作
+ 通过with语句方便的使用上下文管理器，支持上下文管理器协议的对象，像文件就可以使用with语句

> 使用语法： 把上下文管理表达式的值赋给变量，然后执行语句序列，比如下例：

```python
with open('test_tweets.csv',encoding='utf8') as f:
    lista=[]
    for line in f:
        lista.append(line)
    print(lista[1:4])
```

* 上例中打开文件，将文件对象赋给变量f，执行with语句的for循环
* 相比try-finally语句，with语句更加精简和健壮，所以在Python中操作文件和数据库等等对象的时候一般推荐使用with语句

## 用python获取本地数据

* 打开文件

 + 为什么需要关闭文件？
  + 因为python可能会缓存写入的数据，如果程序异常崩溃了，那么数据可能就不会写入到文件中，所以为了安全起见，文件用完以后要主动关闭文件，关闭文件其实就是将缓存写入文件，使得对文件的修改真正起作用

### 打开文件

> file_obj=open(filename,mode='r',buffering=-1……)
+ mode为可选参数，默认为r （read） 还可以有w （write 清空一个文件或者是新建一个文件）
  + a 模式是追加，在文件的尾部追加内容
  + r+=r+w（以读写模式打开）
  + w+=w+r（以读写模式打开，清空原内容）
  + a+=a+r（以读和追加模式打开）
  + 这些模式后面+b的话就代表二进制文件的读写追加
+ buffering也为可选参数，默认-1（表示使用系统默认的缓冲区大小）（0代表不缓冲，1或大于1的值代表缓冲1行或者指定的缓冲区大小）
  + python文件中 二进制文件可以不使用缓冲，但是文本文件必须使用缓冲

** 列表解析 **

> 列表解析的基本语法是:<br>
[ expression for expr in sequence1<br>
　　　　　　　for expr2 in sequence2<br> 
　　　　　　　for expr3 in sequence3<br> 
　　　　　　　if condition
]
* 是由多个for循环以及可迭代的序列构成
* 没有if条件的执行流程
 + 每次把sequence中的一个元素放到expr中，然后再在expression中应用这样的一个对象，形成一个列表
* 有if条件的执行流程
 + 只有满足if条件的内容，才把sequence中的相应内容放到expr中去，再在表达式里应用这样的一个对象，形成一个列表
* 中文输入法下 切换到全角 按空格就可以生效 点 标点符号那个选项，右击 ，全半角，红色圆圈代表现在在全角状态，空格有效

```python
[x**2 for x in range(10)]
```

```python
[x**2 for x in range(10) if x**2<50]
```

** 生成器表达式 **

> 通过圆括号创建生成器表达式（Generator expression） <br>
 生成器表达式并不是创建一个列表，而是返回一个生成器，<br>
 这个生成器在每次计算出一个条目以后，才把这个条目产生出来
 
 生成器表达式并不真正的创建数字列表，而是返回一个生成器对象，此对象在每次计算出一个条目后，把这个条目"产生"(yield)出来。生成器表达式使用了"惰性计算"或称作"延时求值"的机制。 在大数据量计算时，有优势

　　序列过长，并且每次只需要获取一个元素时，应该考虑生成器表达式而不是列表解析。

语法：

　　(expression for iter_val in iterable)

　　(expression for iter_val in iterable if cond_expr)

```python
(x for x in range(10))
```

```python
N=(x for x in range(10))
```

```python
next(N)
```

```python
for i in N:
    print(i)
```

## 聚类分析

聚类是数据挖掘描述性任务和预测性任务的一个重要组成部分，它以相似性为基础，把相似的对象通过静态分类，分成不同的组别和子集。

### ** K-均值**
* 聚类算法有很多，其中k-均值因为其快速，简洁，所以使用广泛，一般认为K-均值算法简单，但是够用
* 基本流程：
    + 任意选择k个对象，作为初始的聚类中心
    + 对每个点，确定它的聚类中心点（其实就是计算距离，一般采用均方差作为标准测度函数）
    + 再计算每个新聚类的计算中心，直到收敛
* 保证各聚类内尽可能紧凑，各聚类之间尽可能分散
* k-means算法不是全局最优解，而是局部最优解，由于初试聚类点选择的随机性，所以聚类的结果可能会发生变化

** 一般的聚类算法包中都会包含k-均值这样的基础算法 **
+ 例如，Scipy中的cluster模块有一个vq，这是一个矢量量化包，其中就包含k-means这样的算法
+ 下面自己创了个数据，做聚类实验

```python
name=["小明","大明","小鹏","大鹏","小萌","大萌"]
```

```python
grades=pd.DataFrame(index=name)
```

可以先创立一个空DataFrame，然后再填数据

```python
import numpy as np
from scipy.cluster.vq import vq,kmeans,whiten
```

```python
list1=list(grades.iloc[0,:].values)
list2=list(grades.iloc[1,:].values)
list3=list(grades.iloc[2,:].values)
list4=list(grades.iloc[3,:].values)
list5=list(grades.iloc[4,:].values)
list6=list(grades.iloc[5,:].values)
list1
# 除了老师给的一个一个输入之外，也可以直接提取
```

```python
data=np.array([list1,list2,list3,list4,list5,list6])
data
```

```python
np.vstack((grades['高数'],grades['英语'],grades['Python'],grades['音乐']))
```

```python
np.vstack((grades.iloc[0,:],grades.iloc[1,:],grades.iloc[2,:],grades.iloc[3,:],grades.iloc[4,:],grades.iloc[5,:]))
# 这样提取才是最快的，就是不知道里面是不是可以写成更简单的形式
```

```python
whitenone=whiten(data)
```

```python
centroids,_=kmeans(whitenone,2)
```

```python
result,_=vq(whitenone,centroids)
```

```python
print(result)
```

* 解释：
    * whiten 算出各列元素的标准差，返回一个新的数组(求点之间的距离，一般使用标准差（均方差）作为度量标准)
    * 聚类的核心函数是 kmeans和vq
        + kmeans函数的功能是对数据进行聚类，两个参数，一个是数据，一个是要分成几类，我们在找学霸，那么就是有潜力成为学霸的（和大萌 第六个元素 一组的） 没潜力成为学霸的（不和大萌一组的）
        + kmeans函数的返回结果有两个  
            + codebook：ndarray k个由k个质心组成的数组。聚类中心数组。 生成的质心和代码表示所见的最低失真，不一定是全局最小失真。
    
            + distortion : float失真：浮动 通过的观测值与生成的质心之间的失真。
        + controids用来接收聚类中心数组，剩下的误差 使用_ 接收，表示我们不使用这个返回结果
        + vq 矢量量化函数，可以对每个数据（也就是这里每个人的成绩）进行归类，然后获得结果
