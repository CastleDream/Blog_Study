## shared 变量

定义shared value

```python
np.array(0,dtype=np.float64)
```

```python
state=theano.shared(np.array(0,dtype=np.float64),name='state')
inc=T.scalar('inc',dtype=state.dtype) #不能直接用np.dtype('float64')
accumulator=theano.function([inc],state,updates=[(state,state+inc)])
```

```python
print(accumulator(10)) #每次运行这个，每次在上面的结果上加10
#但是直接print 得到的state是没有更新之前的，
print(state.get_value())
#所以最科学正确的查看shared value的方式是用get_value
```

设置shared value的值

```python
tmp_func=state*2+inc
a=T.scalar(dtype=state.dtype) #dtype一定要一样
skip_shared=theano.function([inc,a],tmp_func,givens=[(state,a)])
'''
可以看到，tmp_func的操作是在state上进行的，但是在写function的时候，givens
中给了参数，用a来代替state，这样就不会影响state了
'''
print(skip_shared(2,3))
print(state.get_value())
```

## 自定义层

```python
'''
to define the layer like this:
l1=Layer(inputs,in_size=1,out_size=10,activation_function)
l2=Layer(l1.outputs,in_size=10,out_size=2,activation_function=None)
'''

class Layer(object):
    def __init__(self,inputs,in_size,out_size,activation_function=None):
        self.W=theano.shared(np.random.normal(0,1,(in_size,out_size)))
        '''
        这里使用高斯分布，均值为0，方差为1，是为了让初始时候的权重尽可能
        混乱，这样比较有利于之后的更新，
        如果全部为0的话，可能会使网络产生一个不良的结果
        '''
        self.b=theano.shared(np.zeros((out_size,))+0.1)
        #一个1d的，保证bias是一个正数
        self.Wx_plus_b=T.dot(inputs,self.W)+self.b
        # x就是input 
        self.activation_function=activation_function
        if activation_function is None:
            self.outputs=self.Wx_plus_b
        else:
            self.outputs=self.activation_function(self.Wx_plus_b)
        
```

```python
#determine The inputs dtyoe
x=T.dmatrix('x') #  String form:    TensorType(float64, matrix)
y=T.dmatrix('y')

#add layer
l1=Layer(x,1,10,activation_function=T.nnet.relu)
'''
因为这里制造的假数据x_data只有一个维度，所以in_size=1,out_size也就是
本层神经元个数定为10
'''
l2=Layer(l1.outputs,10,1,None)
'''
这里制造的假数据的y_data也只有一个维度，所以这里l2是输出层的话，out_size=1
'''
# compute the cost
cost=T.mean(T.square(l2.outputs-y))

# compute the gradients 
gW1,gb1,gW2,gb2=T.grad(cost,[l1.W,l1.b,l2.W,l2.b])
'''
计算梯度，这里有两层，每层需要计算的其实就是Weight和bias，所以分别写出来
grad参数说明
wrt : :class:`~theano.gof.graph.Variable` or list of Variables
    Term[s] with respect to which we want gradients
'''
# apply gradient descent
learning_rate=0.05
train=theano.function(inputs=[x,y],
                     outputs=cost,
                     updates=[
                         (l1.W,l1.W-learning_rate*gW1),
                             (l1.b,l1.b-learning_rate*gb1),
                             (l2.W,l2.W-learning_rate*gW2),
                             (l2.b,l2.b-learning_rate*gb2)
                             ]
                     )
# prediction
predict=theano.function(inputs=[x],outputs=l2.outputs)

#真正的运算，上面都是定义结构，下面才开始放数据
for i in range(1000):
    err=train(x_data,y_data)
    if i%50==0:
        print(err)
```

## classification例子

```python
def compute_accuracy(y_target,y_predict):
    correct_prediction=np.equal(y_predict,y_target)
    '''
    Returns
out : ndarr ay or scalar
    Output array, element-wise comparison of `x1` and `x2`.
    Typically of type bool,
    '''
    accuracy=np.sum(correct_prediction)/len(correct_prediction)
    #分对的个数/总样本个数
    return accuracy
```

```python
# generate fake data
rng=np.random

N=400 #traing数据个数
feats=784  #input的feature个数

#生成随机数
#D=(input_values,targer_class)
#() 括号代表元组，D[0]就是数据集，D[1]就是数据集对应的标签集
D=(rng.randn(N,feats),rng.randint(size=N,low=0,high=2))
```

```python
np.random.randn(4,3),np.random.randint(size=4,low=0,high=2)
'''
randn:Return a sample (or samples) from the "standard normal" distribution.
randint:Return random integers from `low` (inclusive) to `high` (exclusive). 
'''
```

```python
# declare theano symbolic variables
x=T.dmatrix('x')
y=T.dvector('y')
'''
不过上面的regression例子y用的也是dmatrix
'''

#initialize the weights and biases
W=theano.shared(rng.randn(feats),name='w')
b=theano.shared(0.1,name='b')
'''
上个例子中这里用的是自定义一个Layer，这里不像使用稍微复杂点的层，直接搞
'''
#construct theano expression graph
p_1=T.nnet.sigmoid(T.dot(x,W)+b)
'''
把y=Wx+b 传给激活函数sigmoid激活一下，认为概率大于0.5就是True
'''
prediction=p_1>0.5
# cross_entropy 差不多可以等同于上面例子的cost
'''
https://en.wikipedia.org/wiki/Cross_entropy
这里有cross_entropy的介绍，公式差不多就是这样，
'''
xent=-y*T.log(p_1)-(1-y)*T.log(1-p_1) #对每一个sample的
'''
也可以写成
xent=T.nnet.binary_crossentropy(p_1,y)
'''
cost=xent.mean()+0.01*(W**2).sum() 
#为了克服overfitting的问题，这里在xent_mean()后面加上一个l2正则化 
'''
https://blog.csdn.net/jinping_shi/article/details/52433975
l1正则化和l2正则化
theano中，dropout需要自己编，比较麻烦
'''
gW,gb=T.grad(cost,[W,b])

#compile
learning_rate=0.05
train=theano.function(inputs=[x,y],
                     outputs=[prediction,xent.mean()],
                     updates=[(W,W-learning_rate*gW),
                             (b,b-learning_rate*gb)]
                     )
predict=theano.function(inputs=[x],outputs=prediction)
#training
for i in range(500):
    pred,err=train(D[0],D[1])
    if i%50==0:
        print('cost:',err)
        print('accuracy:',compute_accuracy(D[1],predict(D[0])*1) )
print('target values for D:')
print(D[1])
print('prediction on D:')
print(predict(D[0])*1)
#这里之所以加了一个乘1，是因为predict矩阵本来是bool，乘1之后，True的元素变成1，False变成0，不然没法计算accuracy
```

## Regularization正则化

```python
# 导入包
import theano 
from sklearn.datasets import  load_boston
import theano.tensor as T
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 数据集查看
np.random.seed(100)
x_data=load_boston().data
y_data=load_boston().target[:,np.newaxis] #也变成矩阵
```

```python
x_data.shape,y_data.shape
'''
一共506个数据，每个数据13个特征，波士顿房价预测，回归问题
'''
```

```python
adata=np.array(
    [[1,2],
      [3,4],
      [5,3]])
print(np.max(adata,axis=0))
'''
If `axis` is given, the result is an array of dimension
    ``a.ndim - 1``. 这句解释感觉有问题，不对
ndim讲解： https://www.zhihu.com/question/64894713
shape中有几个数字，ndim就是几（需要几个数字才能索引到一个array最内层的元素，ndim就是几）
'''
print(adata.ndim)
```

```python
def minmax_normalization(data):
    '''
    x 因为各个 feature 的取值范围区别较大，
    所以用 minmax_normalization 对数据进行归一化，
    这样可以把每个 feature 都压缩到 0-1 的范围。
    '''
    xs_max=np.max(data,axis=0)   #所以这个返回的是13个特征列中，每列特征的最大值
    xs_min=np.min(data,axis=0)
    xs=(1-0)*(data-xs_min)/(xs_max-xs_min)+0
    return xs
```

```python
x_data=minmax_normalization(x_data)
```

```python
# 分成训练集测试集

x_train,y_train=x_data[:400],y_data[:400]
x_test,y_test=x_data[400:],y_data[400:]

x=T.dmatrix('x')
y=T.dmatrix('y')
```

```python
class Layer(object):
    def __init__(self, inputs, in_size, out_size, activation_function=None):
        self.W = theano.shared(np.random.normal(0, 1, (in_size, out_size)))
        self.b = theano.shared(np.zeros((out_size, )) + 0.1)
        self.Wx_plus_b = T.dot(inputs, self.W) + self.b
        self.activation_function = activation_function
        if activation_function is None:
            self.outputs = self.Wx_plus_b
        else:
            self.outputs = self.activation_function(self.Wx_plus_b)
    
l1=Layer(x,13,50,T.tanh)
l2=Layer(l1.outputs,50,1,None)
```

```python
#定义cost等
# cost=T.mean(T.square(l2.outputs-y)) #without regularization

'''
下面两个分别是with l2 regularization和with l1 regularization
'''
# cost=T.mean(T.square(l2.outputs-y))+0.1*((l1.W**2).sum()+(l2.W**2).sum())
cost=T.mean(T.square(l2.outputs-y))+0.1*(abs(l1.W).sum()+abs(l2.W).sum())

#先看不使用任何正则化的
# 然后再去看使用l2和l1的
gW1,gb1,gW2,gb2=T.grad(cost,[l1.W,l1.b,l2.W,l2.b])
# 定义学习率 训练函数等
learning_rate=0.01
train=theano.function(inputs=[x,y],
                     updates=[(l1.W,l1.W-learning_rate*gW1),
                             (l1.b,l1.b-learning_rate*gb1),
                             (l2.W,l2.W-learning_rate*gW2),
                             (l2.b,l2.b-learning_rate*gb2)
                             ]
                     )
compute_cost=theano.function(inputs=[x,y],
                            outputs=cost)

# 训练模型
train_err_list=[]
test_err_list=[]
learning_time=[]

for i in range(1000):
    train(x_train,y_train)
    if i%10==0:
        train_err_list.append(compute_cost(x_train,y_train))
        test_err_list.append(compute_cost(x_test,y_test))
        learning_time.append(i) 
```

```python
plt.plot(learning_time,train_err_list,'r-')
plt.plot(learning_time,test_err_list,'b--')
# 加了l1正则的
```

```python
plt.plot(learning_time,train_err_list,'r-')
plt.plot(learning_time,test_err_list,'b--')
# 加了l2正则的
```

```python
# 可视化结果 没有正则化的
plt.plot(learning_time,train_err_list,'r-')
plt.plot(learning_time,test_err_list,'b--')
```

+ 没有正则化时，训练集的误差会逐渐变小，但是测试集的误差越来越大，这就是有 overfitting 。
+ 加入l2正则后，用 L2 正则化时，比上一个好很多，测试集的误差也在不断减小。
+ 加入l1正则后，反正都比不加要好，过拟合现象还是比较严重的,l2防止过拟合，
+ 关于l1正则和l2正则的解释： https://www.zhihu.com/question/20700829

## 保存模型

https://morvanzhou.github.io/tutorials/machine-learning/theano/3-6-save/ 直接看代码好了

```python
import numpy as np
import theano
import theano.tensor as T
import pickle

def compute_accuracy(y_target, y_predict):
    correct_prediction = np.equal(y_predict, y_target)
    accuracy = np.sum(correct_prediction)/len(correct_prediction)
    return accuracy

rng = np.random

# set random seed
np.random.seed(100)

N = 400
feats = 784

# generate a dataset: D = (input_values, target_class)
D = (rng.randn(N, feats), rng.randint(size=N, low=0, high=2))

# Declare Theano symbolic variables
x = T.dmatrix("x")
y = T.dvector("y")

# initialize the weights and biases
w = theano.shared(rng.randn(feats), name="w")
b = theano.shared(0., name="b")

# Construct Theano expression graph
p_1 = 1 / (1 + T.exp(-T.dot(x, w) - b))
prediction = p_1 > 0.5
xent = -y * T.log(p_1) - (1-y) * T.log(1-p_1)
cost = xent.mean() + 0.01 * (w ** 2).sum()
gw, gb = T.grad(cost, [w, b])

# Compile
learning_rate = 0.1
train = theano.function(
          inputs=[x, y],
          updates=((w, w - learning_rate * gw), (b, b - learning_rate * gb)))
predict = theano.function(inputs=[x], outputs=prediction)

# Training
for i in range(500):
    train(D[0], D[1])
```

```python
#这里需要自己新建一个save文件
with open('save/model.pickle', 'wb') as file:
    '''
    保存权重w和偏置b
    '''
    model = [w.get_value(), b.get_value()]
    pickle.dump(model, file)
    print(model[0][:10]) #这打印出的是模型的w
    print('accuracy:',compute_accuracy(D[1],predict(D[0])))
```

```python
with open('save/model.pickle', 'rb') as file:
    model = pickle.load(file)
    '''
    加载已经训练好的w和b
    '''
    w.set_value(model[0])  
    #之前定义的model是个列表，其中列表的第一个元素整体是w，第二个是b
    b.set_value(model[1])
    print(w.get_value()[:10]) #这打印出的是模型的w
    print('accuracy:',compute_accuracy(D[1],predict(D[0])))
```

### axis=0和1的问题

参考这个链接： https://zhuanlan.zhihu.com/p/30960190<br>

**最直白的理解：最外面的括号代表着 axis=0，依次往里的括号对应的 axis 的计数就依次加 1**

![img](https://pic4.zhimg.com/80/v2-93d8cd8c8ae6745394150a7c5f5ed663_hd.jpg)

上图说明，axis 表示 那个级别的括号内的元素 对应位置相加.

> 详细说明：<br>
+ 当axis=0时，[[0,1],[2,3]] 就等于[0,1]+[2,3] 相当于是说 把那层括号里的东西，只要是逗号隔开的，都当一个整体，一个元素去进行。 
+ 当axis=1时，同理，[0]+[1] [2]+[3]

### Theano.dot()讲解

T.dot(a, b)
Docstring:
Computes the dot product of two variables.（计算两个变量的点积）

For two matrices, this is equivalent to matrix multiplication.
For two vectors, this is the inner product.
When one variable is a scalar, this is like elementwise multiplication.
For N dimensions, this is a sum product over the last axis
of the first array and the second-to-last axis of the second array:

    dot(a, b)[i,j,k,m] = sum(a[i,j,:] * b[k,:,m])

Note that this dot function does one of three things, in the following
sequence:

    1.  If either a or b is scalar, it returns the elementwise product
        without calling the Theano Dot op.

    2.  If either a or b has more than 2 dimensions, it calls Theano's
        tensordot function with appropriate axes. The tensordot function
        expresses high-dimensional dot products in terms of 2D matrix
        multiplications, so it may be possible to futherize optimize for
        performance.

    3.  If both a and b have either 1 or 2 dimensions, it calls Theano's
        Dot op on a and b.

Notes

Matrix-matrix products are sometimes optimized to Dot22 or Gemm ops
(see tensor.blas).<br>
Vector-vector products are sometimes optimized to Ger or CGer (see
tensor.blas).<br>
Matrix-vector products are sometimes optimized to Gemv, CGemv (see
tensor.blas).<br>
